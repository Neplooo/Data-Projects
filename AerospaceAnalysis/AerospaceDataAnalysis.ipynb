## Background Information

When flying Airplanes, weather is one of the most crucial things that could make the difference between a successful flight and a deadly crash. If you look at many of the crashes in general aviation, you will see that most of them are due to a lack of attention to weather circumstances in the preflight protocol.

## How to read a METAR Code

Let's Look at this example:

METAR EHLE 280925Z AUTO 21009G19KT 060V130 5000 -RA FEW007 BKN014CB BKN017 02/M01 Q1001 BECMG 6000

Basic terms

*   The "METAR" tag will specify whether the code is instrument based or observation based. If it is observation based, it will show SPECI.
*   The next word is the ICAO Code or the identifier code of the airport.
*   The third value is the date and time in a special form.
*   The fifth and sixth values are the wind information.
*   The seventh value shows the visibility in meters.
*   The eight value shows the weather condition.
*   The 9, 10, and 11 values show the cloud condition.
*   the 12 value shows the temperature.

## Time V Temp
Observing temprature patterns over time is very important. As part of our first test, we will individually graph out the temperature at both the airports over the entire 96 hours that we have data for. Keep in mind that the graph starts at 2:00 PM On 11/25 and finisher at 1:00 AM on 11/29.

```python
fig, ax = plt.subplots()

ax.plot(KGMUData["reportTime"], KGMUData["temp"])

ax.set_title("Time V Temp @ KGMU")
ax.set_ylabel("Temp (Celsius)")
ax.set_xlabel("Date")

plt.xticks(rotation=25)

#ax.xaxis.set_major_formatter(ticker.NullFormatter())
#ax.xaxis.set_major_locator(ticker.NullLocator())
```

```python
fig, ax = plt.subplots()

ax.plot(KCHSData["reportTime"], KCHSData["temp"])

ax.set_title("Time V Temp @ KCHS")
ax.set_ylabel("Temp (Celsius)")
ax.set_xlabel("Date")

plt.xticks(rotation=25)

#ax.xaxis.set_major_formatter(ticker.NullFormatter())
#ax.xaxis.set_major_locator(ticker.NullLocator())
```

### Results

Looking at this first set of data, we can tell that in general, Charleston's airport has a warmer climate than Greenville's airport. As we all know, this is due to Charleston being further down south and closer to the ocean and equator than Greenville, leading to a warmer climate. You can also see where the peaks and drops are in the graph. Those peaks and drops correspond with day and night as those are when temperatures are highest and lowest.

## Cloud Coverage

Here, we will look at the cloud coverage of the two airports and see a difference in cloud patterns.


```python
fig, ax = plt.subplots(1, 2, squeeze=True)

plt.rc("axes", titlesize=7)

ax[0].hist(KGMUData["clouds/0/cover"], align="mid")

ax[0].set_title("Cloud Coverage Report Types @ KGMU") # Font too big (Change Size?)

ax[0].set_ylabel("Reports")
ax[0].set_xlabel("Types of Codes")

ax[1].hist(KCHSData["clouds/0/cover"], align="mid")

ax[1].set_title("Cloud Coverage Report Types@ KCHS")

#ax[1].set_ylabel("Reports")
ax[1].set_xlabel("Types of Codes")
```

### Results

As you can see, there tends to be more clouds in Charleston than in Greenville. This is most likely due to the more volatile weather of the coast. The winds on the coast tend to push more clouds inland causing the Charleston Airport to be cloudier. The greenvlle area is relatively clear most of the time, however, there seem to be a lot of times that there is overcast in the area.
